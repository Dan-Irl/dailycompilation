```python
from bs4 import BeautifulSoup
import cloudscraper
```

```python
url = "https://ads.tiktok.com/business/creativecenter/inspiration/popular/hashtag/pc/en"

scraper = cloudscraper.create_scraper(
    interpreter='nodejs', 
	delay=10, 
    browser={
        'browser': 'chrome',
        'platform': 'windows',
        'mobile': False
    },
    )
req = scraper.get(url)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```
